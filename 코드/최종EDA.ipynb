# 시각화

## 사고 비율

```python
# 원 그래프 그리기
def func(pct, allvals):
    absolute = int(pct/100.*sum(allvals))
    return "{:.1f}%\n({:d} 건)".format(pct, absolute)

plt.figure(figsize=(10, 10))  # 그림 크기 설정
sizes = df_all['교통사고명'].value_counts()
labels = df_all['교통사고명'].unique()

wedges, texts, autotexts = plt.pie(sizes, labels=labels, autopct=lambda pct: func(pct, sizes), startangle=140, textprops={'fontproperties': fontprop, 'fontsize': 10})

# 라벨을 퍼센트 값 위로 조정
for i, text in enumerate(texts):
    text.set_fontproperties(fontprop)
    autotexts[i].set_text(text.get_text() + "\n" + autotexts[i].get_text())
    text.set_text("")

plt.axis('equal')

# 그래프를 보여줍니다.
plt.title('카테고리 분포', fontproperties=fontprop)  # 한글 제목 설정
plt.show()
```
