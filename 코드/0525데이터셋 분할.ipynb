# 학교

```python
school=pd.read_csv("C:/Users/wjdtj/OneDrive - dgu.ac.kr/데이터 분석/교통사고/0525/전국초중등학교위치표준데이터.csv",encoding='cp949')
school.columns
```

```python
school.head()
```

```python
# '설립일자' 열을 datetime 타입으로 변환
school['설립일자'] = pd.to_datetime(school['설립일자'])
```

```python
# 2017년 1월 1일 이전에 설립된 학교만 필터링
school_14_16 = school[school['설립일자'] <= pd.Timestamp('2017-01-01')]
school_14_16 = school_14_16[(school_14_16['소재지지번주소'].str.startswith('전북특별자치도 전주시', na=False))]
```

```python
# 2020년 1월 1일 이전에 설립된 학교만 필터링
school_17_19 = school[school['설립일자'] <= pd.Timestamp('2020-01-01')]
school_17_19 = school_17_19[(school_17_19['소재지지번주소'].str.startswith('전북특별자치도 전주시', na=False))]
```
